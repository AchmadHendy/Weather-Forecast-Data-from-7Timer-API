This project aims to get weather data from 7Timer website.

We can access the API in this link
https://www.7timer.info/doc.php

I extract weather forecast in Bekasi, Indonesia since i live there.

The description of each column can be found in the link.

There are several types of products that we can choose :
1. CIVIL (for civil users)
2. ASTRO (for astronomical purpose)
3. etc

The data should display 8-days forecast since i use CIVIL product data.

```python
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json

#To access the API just insert the URL that has been given in the website.
#Bekasi location coordinates are : longitude = 106.977347 and latitude = -6.250481
url = 'http://www.7timer.info/bin/api.pl?lon=106.977347&lat=-6.250481&product=civil&output=json'
parameters = {
  'start':'1',
  'limit':'5000',
}
headers = {
  'Accepts': 'application/json',
}

session = Session()
session.headers.update(headers)

try:
  response = session.get(url, params=parameters)
  data = json.loads(response.text)
  print(data)
except (ConnectionError, Timeout, TooManyRedirects) as e:
  print(e)
```
